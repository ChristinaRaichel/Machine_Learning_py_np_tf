```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

Generating best fit line for x and y

```python
np.random.seed(126)
```

```python
x = np.linspace(0,50,100)
y = np.linspace(0,50,100)

x += np.random.uniform(-3,3,100) # add noise using uniform
y += np.random.uniform(-3,3,100)
```

```python
y[:10]
```

```python
def init_params():
    w = tf.Variable(np.random.randn(),name='w') # init using rand normal
    b = tf.Variable(np.random.randn(),name='b') 
    return w,b
```

```python
learning_rate = 0.001
training_epochs = 10000
```

```python
def linear_regression(x,w,b):
    y_predicted = tf.add(tf.multiply(x,w),b) #
    return y_predicted



def loss_function(y_predicted,y):
    loss = tf.reduce_mean(tf.square(y-y_predicted)) #
    return loss
```

```python


w,b = init_params()
        
for epoch in range(training_epochs):
    with tf.GradientTape() as tape: #
        y_prediction = linear_regression(x,w,b)
        loss = loss_function(y_prediction,y)

        gradients = tape.gradient(loss,[w,b]) #
        
        w.assign_sub(learning_rate * gradients[0]) #
        b.assign_sub(learning_rate * gradients[1])

        if epoch % 50 == 0 :
            print(f' Epoch: {epoch}, cost: {loss.numpy()}, w: {w.numpy()}, b: {b.numpy()}') #

        
```
